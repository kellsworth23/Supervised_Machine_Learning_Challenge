```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
train_df
```

```python
test_df
```

```python
# Convert categorical data to numeric and separate target feature for training data
y_train = train_df["target"]
X_train = train_df.drop(columns = ["target"])
X_train = pd.get_dummies(X_train)
X_train.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
y_test = test_df["target"]
X_test = test_df.drop(columns = ["target"])
X_test = pd.get_dummies(X_test)
X_test.head()
```

```python
# add missing dummy variables to testing set
for column in X_train.columns:
    if column not in X_test.columns:
        print(f'adding column {column}')
        X_test[column] = 0
```

    My prediction before creating, fitting, and scoring the models, and after looking at both of the datasets is that the Logistic Regression model will generate better results. 
    Even though both models used here (Logistic Regression, and Random Forest models) can be used on classification problems, with over 91 variables on this datasets, the Logistic model would perform better, as Random Forest is more for simpler classification problems.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
logisticRegr = LogisticRegression(
    solver='lbfgs', 
    max_iter=200,
    random_state=0).fit(X_train, y_train)
print('Logistic Regression train score: ', logisticRegr.score(X_train, y_train))
print('Logistic Regression test score: ', logisticRegr.score(X_test, y_test))
```

```python
# Train a Random Forest Classifier model and print the model score
randomForestClass = RandomForestClassifier(random_state=0)
randomForestClass.fit(X_train, y_train)
print("RandomForestClassifier train score: ", randomForestClass.score(X_train,y_train))
print("RandomForestClassifier test score: ", randomForestClass.score(X_test,y_test))
```

     After creating, fitting, and scoring the models, I was surprised that the Random Forest model performed better. This could be due to the fact that we are overfitting the model, or due to the fact that the data going into those models was never scaled, which is an important part in preprocessing. 

```python
# Scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

    After scaling the data, and before re-fitting the models on the scaled data, my prediction remains the same. I still think that the Logistic Regression model will generate better results (for the same reasons as before).

```python
# Train the Logistic Regression model on the scaled data and print the model score
logisticRegr_scaled = LogisticRegression(
    solver='lbfgs',
    max_iter=200,
    random_state=0
)
model = logisticRegr_scaled.fit(X_train_scaled, y_train)
print("LogisticRegression scaled train score: ", logisticRegr_scaled.score(X_train_scaled, y_train))
print("LogisticRegression scaled test score: ", logisticRegr_scaled.score(X_test_scaled, y_test))
```

```python
predictions = model.predict(X_test_scaled)
confusion_matrix(y_test, predictions)
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
randomForestClass_scaled = RandomForestClassifier(random_state=0)
model = randomForestClass_scaled.fit(X_train_scaled, y_train)
print("RandomForestClassifier scaled train score: ", randomForestClass_scaled.score(X_train_scaled, y_train))
print("RandomForestClassifier scaled test score: ", randomForestClass_scaled.score(X_test_scaled, y_test))
```

```python
predictions = model.predict(X_test_scaled)
confusion_matrix(y_test, predictions)
```

    Once the data was scaled, and after re-fitting the models on the scaled data, the Logistic Regression was the better model. The Random Forest model remains overfitted, and even though its accuracy improved a tad (from 0.6333 to 0.6344), the Logistic Regression outperfomed it, going from 0.5255 to 0.7603.
    I still wouldn't feel comfortable using this model (accuracy of 0.76) to predict the credit risk of loans.  Specially after seeing the confusion matrix (on the Logistic Regression), ~11.48% of the testing data were false positives, and ~12.48% were false negatives, ~37.51% where true positives, and ~38.51% were true negatives. Money would be lost if we use this model for loan decisions.
